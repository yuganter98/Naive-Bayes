```python
from sklearn.naive_bayes import CategoricalNB
clf=CategoricalNB()
```

```python
import pandas as pd
NB=pd.read_csv('D:\\Assignment Excelr\\Naiye base\\SalaryData_Test.csv')
NB
```

```python
NB.drop(NB.columns[[1,2,4,5,6,7,8,12]], axis=1, inplace=True)
NB
```

```python
NB_label=NB.copy()
NB_label
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder=LabelEncoder()
NB_label.iloc[:,-1]=labelencoder.fit_transform(NB_label.iloc[:,-1])
NB_label.iloc[:,0]=labelencoder.fit_transform(NB_label.iloc[:,0])
NB_label.iloc[:,1]=labelencoder.fit_transform(NB_label.iloc[:,1])
NB_label.iloc[:,2]=labelencoder.fit_transform(NB_label.iloc[:,2])
NB_label.iloc[:,3]=labelencoder.fit_transform(NB_label.iloc[:,3])
```

```python
NB_label
```

```python
X=NB_label.iloc[:,:-1]
Y=NB_label.iloc[:,-1]
```

```python
clf.fit(X,Y)
```

```python
import numpy as np
X_test=np.array([[2,0,0,1,0]])
X_test.shape
X_test.reshape(1,-1)
```

```python
print(clf.predict(X_test))
```
